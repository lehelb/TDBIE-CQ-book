```python
import numpy as np
import cq_lms #Import library of functions for CQ based on linear multistep methods
import matplotlib.pyplot as plt
```

```python
# Data
g = lambda t : np.sin(t)**2
def gd(t):
    I = t < 0
    res = 2*np.cos(t)*np.sin(t)
    res[I] = 0
    return res
```

```python
# Hyperbolic kernel
Ks = lambda s: s/(1-np.exp(-s))
```

```python
T = 5
N = 300
# cq_lms
u = cq_lms.evalCQ(g,Ks,N,T,cq_lms.BDF2)
```

```python
ts = np.linspace(0,T,N+1)
```

```python
plt.plot(ts,u.real)
plt.plot(ts,gd(ts)+gd(ts-1)+gd(ts-2)+gd(ts-3)+gd(ts-4))
```

```python
plt.plot(ts,gd(ts)+gd(ts-1)+gd(ts-2))
```
